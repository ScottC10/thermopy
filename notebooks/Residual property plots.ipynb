```python
from thermopy.properties import residual_enthalpy
from thermopy.eos import PengRobinson
import numpy as np
import matplotlib.pyplot as plt
```

```python
PR = PengRobinson("methane")
T_range = [300, 500, 800, 1500, 3000]
p_range = np.linspace(0.01, 50*101325, 5000)
Hrs = []

for t in T_range:
    Hrs = []
    for p in p_range:
        state = PR.solve(t, p)
        state=state[-1]

        Hr = residual_enthalpy(state, PR)
        Hrs.append(Hr)

    plt.plot(p_range, Hrs, label=f"T = {t}")
plt.xlabel("pressure [Pa]")
plt.ylabel("residual enthalpy J/mol")
plt.legend()
plt.show()
```
